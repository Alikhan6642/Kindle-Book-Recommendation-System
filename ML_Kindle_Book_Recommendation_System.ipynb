## <font color='#633974'><u>About the Function: `missing_values`</u></font>
<font color='#117A65' size='4'>The `missing_values` function is designed to analyze the missing values in a DataFrame</font>


```python
def missing_values_function(df):
    # Count missing values in each column
    missing_value_count = df.isnull().sum()

    # Calculate the percentage of missing values in each column
    missing_value_percentage = round(df.isnull().mean() * 100, 2)

    # Combine missing value count and percentage into a single table
    missing_data_table = pd.concat([missing_value_count, missing_value_percentage], axis=1)

    # Rename columns for better clarity
    missing_data_table = missing_data_table.rename(
        columns={missing_data_table.columns[0]: 'Missing Values', missing_data_table.columns[1]: '% of Total Values'}
    )

    # Add data type of each column
    missing_data_table['Data Type'] = df.dtypes

    # Sort the table by percentage of missing values in descending order
    missing_data_table = missing_data_table.sort_values('% of Total Values', ascending=False)

    # Reset the index for cleaner output
    return missing_data_table.reset_index()
```

```python
# Call the function to check for missing values in the 'users' DataFrame
missing_values_function(users)
```

```python
# Fill missing values in the 'Age' column by imputing the median age of each country group.
# The 'groupby' function groups the data by 'Country', and 'transform' applies the median
# age of each country group to the corresponding rows with missing 'Age' values.
users['Age'] = users['Age'].fillna(users.groupby('Country')['Age'].transform('median'))
```

<font color='#117A65' size='4'>A recommender system aims to predict a user's preferences for various items by analyzing past interactions. Two widely-used techniques for building such systems are Content-Based Filtering and Collaborative Filtering.

<font color='#117A65' size='4'>Collaborative filtering, commonly employed by platforms like Amazon, YouTube, and Netflix, suggests items to users by leveraging the preferences of similar users. This method can be divided into two main categories: memory-based and model-based collaborative filtering.</font>

<font color='#117A65' size='4'>The model-based approach focuses on applying machine learning algorithms to forecast user ratings. It typically uses dimensionality reduction techniques to address the challenge of sparse, high-dimensional data. By transforming a large, sparse matrix into a compact, lower-dimensional representation, these algorithms can make more accurate predictions despite the presence of many missing values.</font>

<font color='#117A65' size='4'>In this section, we will compare the performance of two popular algorithms—Singular Value Decomposition (SVD) and Non-negative Matrix Factorization (NMF). We will experiment with various configurations of their parameters and analyze the results.</font>

<font color='#633974' size='4'>This study aims to develop a book recommendation system utilizing the Book-Crossing dataset. To enhance computational efficiency and minimize memory-related issues, the analysis will focus on users who have submitted at least three ratings and on the top 10% of books with the highest frequency of ratings. The recommendation models will be constructed using the Surprise library, specifically leveraging matrix factorization techniques.</font>

<font color='#117A65' size='4'>Comparison of SVD and NMF Models  
Singular Value Decomposition (SVD) and Non-negative Matrix Factorization (NMF) are both matrix factorization methods employed for dimensionality reduction. The Surprise package offers implementations of these algorithms for use in recommendation systems.</font>

```python
# Create a copy of the filtered ratings DataFrame to avoid modifying the original data
df = df_ratings_top.copy()

# Initialize a Reader object to specify the rating scale (from 1 to 10 in this case)
reader = Reader(rating_scale=(1, 10))

# Load the DataFrame into a Surprise Dataset format for model training
# The 'user_id', 'isbn', and 'book_rating' columns are used to construct the dataset
data = Dataset.load_from_df(df[['user_id', 'isbn', 'book_rating']], reader)
```

```python
# Initialize the SVD (Singular Value Decomposition) model
model_svd = SVD()

# Perform cross-validation on the SVD model using 3-fold cross-validation
# This will evaluate the model's performance on the dataset
cv_results_svd = cross_validate(model_svd, data, cv=3)

# Convert the cross-validation results to a DataFrame and calculate the mean of each metric (e.g., RMSE, MAE)
# This provides an overall performance evaluation of the SVD model
pd.DataFrame(cv_results_svd).mean()
```

```python
# Initialize the NMF (Non-negative Matrix Factorization) model
model_nmf = NMF()

# Perform cross-validation on the NMF model using 3-fold cross-validation
# This evaluates the model's performance on the dataset
cv_results_nmf = cross_validate(model_nmf, data, cv=3)

# Convert the cross-validation results to a DataFrame and calculate the mean of each metric (e.g., RMSE, MAE)
# This gives an overall performance summary of the NMF model
pd.DataFrame(cv_results_nmf).mean()
```

<font color='#117A65' size='4'>The results clearly indicate that, for the given dataset, the SVD (Singular Value Decomposition) approach outperforms NMF (Non-negative Matrix Factorization) in terms of both accuracy and training/testing time. This suggests that SVD is more efficient and yields better performance for this specific task.</font>

# <font color='#2E86C1'>🔧 Optimization of the SVD Algorithm</font>

<font color='#117A65' size='4'>
Grid Search Cross-Validation is a technique that computes accuracy metrics for an algorithm across various combinations of parameters during a cross-validation process. It is an effective method for identifying the optimal configuration of parameters for a given model.

<font color='#117A65' size='4'>In the case of the SVD algorithm, Grid Search is used to find the best settings for the following parameters:</font>
- <font color='#633974'><b>n_factors:</b></font> <font color='#117A65'>The number of latent factors (dimensions) for the matrix factorization.</font><br>
- <font color='#633974'><b>n_epochs:</b></font> <font color='#117A65'>The number of iterations for the Stochastic Gradient Descent (SGD) procedure.</font><br>
- <font color='#633974'><b>lr_all:</b></font> <font color='#117A65'>The learning rate for all parameters, controlling the step size in updating the model.</font><br>
- <font color='#633974'><b>reg_all:</b></font> <font color='#117A65'>The regularization term for all parameters, preventing overfitting by penalizing large weights.</font><br>

After performing Grid Search, it becomes evident that, for most parameters, the default settings already yield the best performance. The improvements gained through Grid Search are minimal, suggesting that the algorithm's default configuration is well-suited for this particular dataset.
</font>


```python
# Defining the parameter grid for Grid Search
param_grid = {'n_factors': [80, 100],              # Number of latent factors for matrix factorization
              'n_epochs': [5, 20],               # Number of iterations for the SGD procedure
              'lr_all': [0.002, 0.005],          # Learning rate for all parameters
              'reg_all': [0.2, 0.4]}             # Regularization term for all parameters

# Setting up GridSearchCV to perform cross-validation with the defined parameter grid
gs = GridSearchCV(SVD, param_grid, measures=['rmse', 'mae'], cv=3)

# Fitting the GridSearchCV to the dataset
gs.fit(data)

# Printing the best RMSE score obtained from the GridSearchCV
print(gs.best_score['rmse'])

# Printing the best parameters that gave the best RMSE score
print(gs.best_params['rmse'])
```

```python
# Check if there are duplicate entries in the dataset based on 'userID' and 'bookTitle'
if not rating_popular_book[rating_popular_book.duplicated(['userID', 'bookTitle'])].empty:
    # Store the initial number of rows in the dataframe
    initial_rows = rating_popular_book.shape[0]

    # Print the initial shape of the dataframe
    print('Initial dataframe shape: {0}'.format(rating_popular_book.shape))

    # Remove duplicate rows based on 'userID' and 'bookTitle'
    rating_popular_book = rating_popular_book.drop_duplicates(['userID', 'bookTitle'])

    # Store the updated number of rows in the dataframe
    current_rows = rating_popular_book.shape[0]

    # Print the new shape of the dataframe and the number of rows removed
    print('New dataframe shape: {0}'.format(rating_popular_book.shape))
    print('Removed {0} duplicate rows.'.format(initial_rows - current_rows))
```

```python
# Create a pivot table to structure ratings data with 'bookTitle' as the index and 'userID' as columns
# Filling missing ratings with 0 to represent unrated books
user_rating_pivot = rating_popular_book.pivot(index='bookTitle', columns='userID', values='bookRating').fillna(0)

# Convert the pivot table into a sparse matrix format using CSR (Compressed Sparse Row) format
# This is done to efficiently store large sparse matrices
user_rating_matrix = csr_matrix(user_rating_pivot.values)

# Optional: Print the shape of the resulting sparse matrix to confirm the structure
print(user_rating_matrix.shape)
```

<font color='#117A65' size='4'>Finding the Nearest Neighbors

```python
# Importing the NearestNeighbors model from sklearn
# This model will be used to find similar books based on user ratings
from sklearn.neighbors import NearestNeighbors

# Initialize the NearestNeighbors model
# We are using 'cosine' similarity metric and 'brute' force algorithm for distance calculations
model_knn = NearestNeighbors(metric='cosine', algorithm='brute')

# Fit the model to the user-item rating matrix
# This will train the model to find the nearest neighbors (similar books) based on the ratings data
model_knn.fit(user_rating_matrix)

# Optional: Print a message confirming that the model is trained
print("KNN model has been trained successfully.")
```

<font color='#117A65' size='4'>Test our model and Make few Recommnedations

```python
# Randomly selecting a query index from the user-item rating matrix
query_index = np.random.choice(user_rating_pivot.shape[0])

# Using the trained KNN model to find the nearest neighbors for the selected query index
# We reshape the input for the model to a 2D array (1, -1) as required by KNN
distances, indices = model_knn.kneighbors(user_rating_pivot.iloc[query_index, :].values.reshape((1, -1)), n_neighbors=6)

# Loop through the neighbors and print out the recommended books
for i in range(0, len(distances.flatten())):
    if i == 0:
        # Print the name of the book for which recommendations are being made
        print('Recommendations for {0}:\n'.format(user_rating_pivot.index[query_index]))
    else:
        # Print each recommended book and its associated distance from the query book
        print('{0}: {1}, with distance of {2}:'.format(i, user_rating_pivot.index[indices.flatten()[i]], distances.flatten()[i]))
```

```python
# Importing necessary libraries for dimensionality reduction
import sklearn
from sklearn.decomposition import TruncatedSVD

# Performing Singular Value Decomposition (SVD) to reduce the matrix to 12 components (dimensions)
SVD = TruncatedSVD(n_components=12, random_state=17)

# Fitting the SVD model and transforming the matrix (X) into a reduced dimensionality space
matrix = SVD.fit_transform(X)

# Checking the shape of the transformed matrix after dimensionality reduction
matrix.shape
```

```python
# Calculating the correlation matrix of the reduced matrix using Pearson correlation coefficient
corr = np.corrcoef(matrix)

# Checking the shape of the correlation matrix
corr.shape
```

<font color='#117A65' size='4'>Let's find books similar to Harry Potter and the Sorcerer's Stone (Book 1)

```python
# # Extracting the list of book titles (columns) from the user rating pivot table
book_title = user_rating_pivot2.columns

# # Converting the book titles to a list
book_list = list(book_title)
```

```python
def recommend_books(book_name):
    # Check if the input book is in the list of book titles
    if book_name not in book_list:
        print(f"Sorry, '{book_name}' is not found in the database. Please try a different book.")
        return

    # Find index of the input book in the list
    book_index = book_list.index(book_name)

    # Get the correlation of this book with all other books
    corr_coffey_hands = corr[book_index]

    # Get the list of recommended books based on correlation
    recommended_books = list(book_title[(corr_coffey_hands < 1.0) & (corr_coffey_hands > 0.9)])

    # Display results
    print(f"\n📚 **Recommended books for '{book_name}':**\n")
    if recommended_books:
        for idx, recommended_book in enumerate(recommended_books[:5], 1):  # Limit to top 5
            print(f"{idx}. {recommended_book}")
    else:
        print("No recommendations available.")

    print("\nHappy reading! 📖✨")

# Example usage
book_input = input("Please Enter a Book Name: ")
recommend_books(book_input)
```
